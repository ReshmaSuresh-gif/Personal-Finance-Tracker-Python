# PROJECT TITLE: PERSONAL FINANCE TRACKER

This project helps users track their income and expenses. It calculates total expenses, income, and remaining balance, and shows a pie chart.

**Pandas** - For handling data,
**Matplotlib** - For visualization

```python
# IMPORTING LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# TAKING EXPENSE DETAILS
empty=[]
while True:
    transaction={"Date":input("Enter the date"),
       "Item Name":input("Enter the Item Name"),
       "Expenses":int(input("Enter the Expenses")),
      }
    empty.append(transaction)
    another = input("Add another? (y/n): ")
    if another == "n":
         break

print(empty)
```

```python
# TAKING INCOME DETAILS
empty1=[] 
while True:
    income=int(input("Enter Your Income"))
    empty1.append(income)
    another = input("Add another? (y/n): ")
    if another == "n":
         break
```

```python
# CALCULATING TOTAL INCOME, EXPENSES, BALANCE
total_income = sum(empty1)
total_expenses = sum(t["Expenses"] for t in empty)
balance = total_income-total_expenses
print(f"Total Income: {total_income}")
print(f"Total Expenses: {total_expenses}")
print(f"Balance: {balance}")
```

```python
# CREATING DATAFRAME
data=pd.DataFrame(empty)
data
```

```python
# DISPLAYING PIE CHART
values = list(data["Expenses"]) + [balance]
labels = list(data["Item Name"]) + ["Income Left"]
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
# LOW BALANCE WARNING
if balance < 2000:
    print("⚠️ Low balance! Only {balance} remaining!")
else:
    print("✅ Balance is healthy!")
```

```python
data.to_csv("finance_data.csv", index=False)
```
